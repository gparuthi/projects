```python
@lview.parallel()
@parallel.require(log)
def count_lines(filep):
    logger = log.logger('Parallel_'+filep)
    f= open(filep)
    line = f.readline()
    count =0
    while line:
        count +=1
        if count%10000==0:
            logger.log('cur_line:'+ count)
        line = f.readline()
    return count

starttime = datetime.now()
res = count_lines.map(input_files)
print 'time taken= '+ str(datetime.now()-starttime)
print res
```
