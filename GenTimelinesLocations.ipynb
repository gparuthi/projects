```python
def gen_time_line_for_loc(loc, locs):
    loc_dict = {}
    for k in locs:
        if loc in locs[k]:
            loc_dict[k] = locs[k][loc]
        else:
            loc_dict[k] = 0
            
    # now generate the histogram
    return getCity(loc, loc_dict)
```

```python
def gen_multiple_timelines_to_graph(loc_list, locs):
    dicts = {}
    for input_loc in loc_list:
        print 'procesing data for: '+input_loc
        loc_dict = {}
        for k in locs:
            date = dateutil.parser.parse(k)
            loc_dict[date] = 0
            for loc in locs[k]:
                if input_loc in loc:
                    loc_dict[date] += locs[k][loc]
        dicts[input_loc] = loc_dict
        
    # now generate the histogram
    return genHistMultiple(dicts)

def calc_diff_from_means(json):
    sum =0
    for date in json:
        sum += json[date]
    avg = sum/len(json)
    ret = {}
    for date in json:
        ret[date] = (json[date]-avg)/avg*100
    return ret

def plot_data(data,loc, graph):
    data.sort()        
    x = [matplotlib.dates.date2num(date) for (date, value) in data]
    y = [value for (date, value) in data]
    # Plot the data as a red line with round markers
    graph.plot(x,y,'-o', label= loc)
    # Set the xtick locations to correspond to just the dates you entered.
    graph.set_xticks(x)
    
    # Set the xtick labels to correspond to just the dates you entered.
    graph.set_xticklabels(
            [date.strftime("%Y-%m-%d-%I") for (date, value) in data],
            rotation=90, size='small'
            )
    graph.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, borderaxespad=0.)

def genHistMultiple(json):
    
    fig = figure(figsize=(len(json[json.keys()[0]])/3,20))
    fig.canvas.set_window_title('multiple cities')
    graph = fig.add_subplot(311)
    graph_means = fig.add_subplot(313)
    for loc in json:
        # generate histogram on this dictionary
        print 'procesing: '+loc
        data = json[loc].items()
        plot_data(data,loc,graph)
        data = calc_diff_from_means(json[loc]).items()
        plot_data(data,loc,graph_means)
        print 'done processing '+loc+'. Now showing...' 

    return graph
```

```python
def getStd(city_dict):
    avg = numpy.std([city_dict[t] for t in city_dict])
    std = numpy.std([city_dict[t] for t in city_dict])
    return std

def get_per_change(city_dict):
    main_hr_count = city_dict[dateutil.parser.parse('2012-11-7 4:0:0')]
    hrsd = []
    for t in city_dict:
        if t.hour == 4 and t != dateutil.parser.parse('2012-11-7 4:0:0'):
            hrsd.append(city_dict[t])
            
    avg = numpy.average(hrsd)
    std = numpy.std(hrsd)
    per_change = (main_hr_count-avg)/avg*100
    return per_change,std

def gen_hist_city_dicts(city_dicts):
    fig = figure()
    fig.canvas.set_window_title('% increase in an hour')
    graph = fig.add_subplot(111)
    final_dict = {}
    for loc in city_dicts:
        print 'procesing: '+loc
        per_change,std = get_per_change(city_dicts[loc])
        final_dict[loc] = {'time':'2012-11-07-04', 'per_change': per_change, 'std':std}
    plot_data_cities_comp(final_dict,graph)
    
"""This method will generate the histogram for input data. i.e. : a json like {city:{time:'', per_change:'', 'std':''}}
"""
def plot_data_cities_comp(data,ax):
    x = data.keys()
    y = [data[loc]['per_change'] for loc in data]
    #Stds= [data[loc]['std'] for loc in data]
    
    ind = numpy.arange(len(x))  # the x locations for the groups
    width = 0.35       # the width of the bars
    
    rects1 = ax.bar(ind, y, width, color='b')
    ax.set_ylabel('% change')
    ax.set_title('% change in tweets for the given hour')
    ax.set_xticks(ind+width)
    ax.set_xticklabels( x )
    
    ax.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3, ncol=2, borderaxespad=0.)   

def get_city_timelines(loc_list, locs):
    dicts = {}
    for input_loc in loc_list:
        print 'procesing data for: '+input_loc
        loc_dict = {}
        for k in locs:
            date = dateutil.parser.parse(k)
            loc_dict[date] = 0
            for loc in locs[k]:
                if input_loc in loc:
                    loc_dict[date] += locs[k][loc]
        dicts[input_loc] = loc_dict
    
    return dicts
```
