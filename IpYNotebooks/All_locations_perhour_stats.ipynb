```python
from ujson import loads
from dateutil.parser import parse
import csv
```

```python
f = open('../DataMining/GardenHose/final_outputs/AllLocs_2013-02-15 12:09:27.783158.data')
json = loads(f.read())
```

```python
def getStd(city_dict):
    avg = numpy.std([city_dict[t] for t in city_dict])
    std = numpy.std([city_dict[t] for t in city_dict])
    return std

def get_per_change(timein, city_dict):
    main_hr_count = city_dict[timein]
    hrsd = []
    for t in city_dict:
        if t.hour == timein.hour and t != timein:
            hrsd.append(city_dict[t])
            
    avg = numpy.average(hrsd)
    std = numpy.std(hrsd)
    abs_change = (main_hr_count-avg)
    per_change = abs_change/avg*100
    return abs_change,per_change,avg,std,main_hr_count

def get_cities_dict(loc_list, locs):
    dicts = {}
    for input_loc in loc_list:
        if len(dicts)%100==0:
            print 'procesing ['+ str(len(dicts))+'/'+str(len(loc_list))+']): '+input_loc
        loc_dict = {}
        for k in locs:
            date = parse(k)
            loc_dict[date] = 0
            for loc in locs[k]:
                if input_loc in loc:
                    loc_dict[date] += locs[k][loc]
        dicts[input_loc] = loc_dict
    
    return dicts
```

```python
import dateutil
def get_city_timelines(loc_list, times):
    dicts = {}
    il_count = 0
    for input_loc in loc_list:
        il_count+=1
        print '[%d/%d] processing data for: %s' %(il_count,len(loc_list),input_loc)
        loc_dict = {}
        for time in times:
            date = dateutil.parser.parse(time)
            loc_dict[date] = 0
            for loc in times[time]:
                if input_loc in loc:
                    loc_dict[date] += times[time][loc]
        dicts[input_loc] = loc_dict
    
    return dicts
# get loc_dict
locs_dict= get_city_timelines(json[json.keys()[0]], json)
```

```python
def write_items_to_file(items, selfpath):
    writer = csv.writer(open(selfpath, 'wb'))
    new_items = [(a.encode('utf-8'),b,c,d,e,f,g) for a,b,c,d,e,f,g in items]
    for item in new_items:
        writer.writerow(item)
        
def write_to_csvs(locs_dict):
    ### generate the final output for each loc
    items =[]
    
    items.append(('location','in_time','per change', 'abs change','avg', 'standard dev', 'main hr count'))
    loc_count =0
    tot_count = len(locs_dict)
    for loc in top:
        loc_count+=1
        print '[%d/%d] processing data for: %s' %(loc_count,tot_count,loc)
        for time in locs_dict[loc]:
            #print time
            time_in = dateutil.parser.parse(str(time))
            abs_change,per_change,avg,std,mhr = get_per_change(time_in,locs_dict[loc])
            if abs_change > std:
                t = (loc,time_in, per_change, abs_change, avg, std, mhr)
                items.append(t)
    
    
        write_items_to_file(items, "../DataMining/GardenHose/final_outputs/stats/stats_"+".csv")
    
    
```

```python
write_to_csvs(locs_dict)
```
