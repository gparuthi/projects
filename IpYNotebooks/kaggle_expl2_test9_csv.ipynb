```python
# sort the locations 
def sort_locs(locs):
    locs_sorted = {}
    for l in locs:
        sorted_teams = [(x,y) for y,x in locs[l].items()]
        sorted_teams.sort()
        sorted_teams.reverse()
        locs_sorted[l] = sorted_teams
    return locs_sorted
    #prefs = " ".join([x.replace('t_','') for y,x in locs_sorted['SZ'][0:10]])
    #prefs
#locs_sorted['US']
locs_sorted = sort_locs(train_locs)
locs_sorted
```
